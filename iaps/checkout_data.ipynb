```python
def show_max_imgs(data,col, num_imgs):
    largest_idx = data.nlargest(num_imgs,col).index
    
    for i in largest_idx:
        img_file = 'IMAGES/'+ str(int(data[1][i])) +'.JPG'
        plt.figure()
        plt.imshow(plt.imread(img_file))
col_num = {'p':2, 'a':4, 'd':6}
show_max_imgs(data,2,5)
```
